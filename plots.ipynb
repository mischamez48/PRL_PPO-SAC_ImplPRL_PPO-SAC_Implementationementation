```python
#graph of best parameters PPO and SAC

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
file_path = "Pendulum_sac_ppo.csv"  # Adjust this path as necessary
data = pd.read_csv(file_path)

# Step 2: Extract columns
scores = data[['scoreSAC', 'scorePPO']]

# Define labels for each score
labels = [
    "actor_lr =3e-3, critic_lr= 3e-3, tau=0.05, alpha=0.2, train_interval=2",
    "actor_lr =5e-4, critic_lr= 3e-4, entropy_coef=0.005",
]

# Step 3: Plot the values
plt.figure(figsize=(10, 6))
for column, label in zip(scores.columns, labels):
    plt.plot(scores.index, scores[column], label=label)

# Add titles and labels
plt.title('Scores Comparison')
plt.xlabel('Episode')
plt.ylabel('Scores')
plt.legend(loc='lower right')  # Move legend to the lower left

# Show plot
plt.show()



```

```python
import pandas as pd

def merge_scores(csv_file1, csv_file2, output_csv):
    # Lire les deux fichiers CSV
    df1 = pd.read_csv(csv_file1, header=None)
    df2 = pd.read_csv(csv_file2, header=None)

    # Fusionner les deux DataFrames en une seule en concaténant les colonnes
    merged_df = pd.concat([df1, df2], axis=1)

    # Écrire le DataFrame fusionné dans un nouveau fichier CSV
    merged_df.to_csv(output_csv, index=False, header=False)

# Exemple d'utilisation
csv_file1 = 'mountain_sac_64_5_1e-3_02_05_128_1_20upe.csv'
csv_file2 = 'score1.csv'
output_csv = 'merged_scores.csv'

merge_scores(csv_file1, csv_file2, output_csv)
```
